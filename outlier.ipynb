-- Box Plot ( Helps to find instantly whether there is outlier in data or not )

**What happens if we have outlier** 

-- Mean and Standard Deviation will be affected 





```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

**Using Z-Score**



 snd.png
        

*   IF the value is above (-,+)3rd quantile range then it is taken as outlier




z-score = Xi - Mean(X) / SD(X)

**z = (X — μ) / σ**



```python
sd = np.std(dataset)
mean = np.mean(dataset)

print(mean,sd)
```

```python
z = []
threshold = 3
outlier = []
for i in dataset:
  value = (i - mean) / sd
  z.append(value)
  if np.abs(value) > 3:
    outlier.append(i)

print(outlier)
```

**Using Box-Plot**

image.png

```python
fig = plt.figure(figsize =(3, 3))
plt.boxplot(dataset)
plt.show()
```

```python
quantile1, quantile3= np.percentile(dataset,[25,75])


iqr_value=quantile3-quantile1


lower_bound_val = quantile1 -(1.5 * iqr_value) 
upper_bound_val = quantile3 +(1.5 * iqr_value)

print(f'Quantile-1 (25%) {quantile1} , Quantile-3 (75%)  {quantile3}')
```

```python
outlier_ = []
m = np.mean(dataset)
SD = np.std(dataset)

for i in dataset:
  if i < lower_bound_val or i > upper_bound_val:
    outlier_.append(i)
print(outlier_)
```
